Complete Sentiment Analysis Function

```python
def sentimentalAnalysis(data, columnname):

    # Downloading VADER Lexicon / Import Required Libraries
    import matplotlib.pyplot as plt
    import seaborn as sns

    "Importing Necessary Package"
    from nltk.sentiment.vader import SentimentIntensityAnalyzer

    # Create Sentiment Analyzer Object
    sid = SentimentIntensityAnalyzer()

    "Deleting null or empty values"
    data.dropna(inplace=True)

    "Checking for a comment"
    # sid.polarity_scores(data[columnname][93])

    "Creating respective columns"

    data['scores'] = data[columnname].apply(
        lambda commentText: sid.polarity_scores(commentText)
    )

    data['compound'] = data['scores'].apply(
        lambda score_dict: score_dict['compound']
    )

    data['Negative'] = data['scores'].apply(
        lambda score_dict: score_dict['neg']
    )

    data['Positive'] = data['scores'].apply(
        lambda score_dict: score_dict['pos']
    )

    data['Neutral'] = data['scores'].apply(
        lambda score_dict: score_dict['neu']
    )

    "Creating final positive or negative using compound score"

    data['comp_score'] = data['compound'].apply(
        lambda c: 'pos' if c >= 0 else 'neg'
    )

    plt.clf()

    # comp = sns.countplot(
    #     x='comp_score',
    #     hue='Name',
    #     data=data,
    #     palette='magma'
    # )

    # comp.figure.savefig("data_charts.png")

    "Checking how many positive and negative messages"

    posneg = pd.DataFrame(data['comp_score'].value_counts())

    return posneg, data
```

A lambda function is a small anonymous (unnamed) function that is written in a single line. It is mainly used when you need a simple function for a short period, especially with functions like apply(), map(), filter(), and sorted().

=======

In WhatsApp sentiment analysis project, lambda is used to process each chat message individually—first to compute sentiment scores, then to extract specific values (like compound, pos, neg, neu), and finally to classify messages as positive or negative.

**Topic Modelling**





Import TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

| Code                              | Description                                          |
| --------------------------------- | ---------------------------------------------------- |
| `from`                            | Imports a specific class from a Python library.      |
| `sklearn.feature_extraction.text` | Module used to convert text into numerical features. |
| `import`                          | Imports the required class.                          |
| `TfidfVectorizer`                 | Converts text documents into TF-IDF feature vectors. |


TF-IDF Vectorizer, which converts text messages into numerical values that machine learning algorithms can understand.

Create TF-IDF Model

```python
tfidf = TfidfVectorizer(
    max_df=0.95,
    min_df=2,
    stop_words='english'
)
```

| Parameter              | Description                                                                            |
| ---------------------- | -------------------------------------------------------------------------------------- |
| `TfidfVectorizer()`    | Creates a TF-IDF vectorizer object.                                                    |
| `max_df=0.95`          | Ignores words appearing in more than **95%** of documents because they are too common. |
| `min_df=2`             | Keeps only words appearing in at least **2 documents**.                                |
| `stop_words='english'` | Removes common English words like **the, is, are, a, an, of**.                         |
| `tfidf=`               | Stores the TF-IDF vectorizer object in the variable `tfidf`.                           |


Creates a TF-IDF model with preprocessing settings to improve topic modeling.

Convert Chat Messages into TF-IDF Matrix

```python
dtm = tfidf.fit_transform(df["Chat"])
```

| Code              | Description                                                                   |
| ----------------- | ----------------------------------------------------------------------------- |
| `df["Chat"]`      | Selects the Chat column from the DataFrame.                                   |
| `fit_transform()` | Learns the vocabulary and transforms the text into TF-IDF values in one step. |
| `dtm`             | Stores the Document-Term Matrix (TF-IDF matrix).                              |


Converts all WhatsApp chat messages into a Document-Term Matrix (DTM).

Import NMF

```python
from sklearn.decomposition import NMF
```

| Code                    | Description                                                          |
| ----------------------- | -------------------------------------------------------------------- |
| `sklearn.decomposition` | Module containing decomposition algorithms.                          |
| `NMF`                   | Non-negative Matrix Factorization algorithm used for Topic Modeling. |


Imports the NMF algorithm to discover hidden topics in the chat messages.

Create NMF Model

```python
nmf_model = NMF(
    n_components=5,
    random_state=42
)
```

| Parameter         | Description                                            |
| ----------------- | ------------------------------------------------------ |
| `NMF()`           | Creates an NMF model.                                  |
| `n_components=5`  | Finds **5 topics** from the chat messages.             |
| `random_state=42` | Produces consistent results each time the code is run. |
| `nmf_model`       | Stores the NMF model.                                  |


Creates an NMF model configured to identify five topics.

| Abbreviation | Expansion                             |
| ------------ | ------------------------------------- |
| **NMF**      | **Non-negative Matrix Factorization** |

====

| Word              | Description                                                                                            |
| ----------------- | ------------------------------------------------------------------------------------------------------ |
| **Non-negative**  | All values in the input and output matrices are zero or positive (no negative numbers).                |
| **Matrix**        | A table of numerical values arranged in rows and columns.                                              |
| **Factorization** | The process of breaking one large matrix into smaller matrices while preserving important information. |


Train the Model

```python
nmf_model.fit(dtm)
```

| Code    | Description                                    |
| ------- | ---------------------------------------------- |
| `fit()` | Trains the NMF model.                          |
| `dtm`   | Uses the TF-IDF Document-Term Matrix as input. |


Learns the hidden topics from the TF-IDF matrix.

Overall Workflow
---------------

| Step | Process               | Output                          |
| ---- | --------------------- | ------------------------------- |
| 1    | Import TF-IDF         | TF-IDF library loaded           |
| 2    | Create TF-IDF object  | Vectorizer ready                |
| 3    | Convert chat messages | Document-Term Matrix (`dtm`)    |
| 4    | Import NMF            | Topic modeling algorithm loaded |
| 5    | Create NMF model      | Model configured for 5 topics   |
| 6    | Fit the model         | Hidden topics learned           |


top 10 words for each topic identified by the NMF Topic Modeling algorithm.

```python
for index, topic in enumerate(nmf_model.components_):
    results = ([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

| Code                            | Description                                                                                                          |
| ------------------------------- | -------------------------------------------------------------------------------------------------------------------- |
| `for`                           | Starts a loop to process each topic one by one.                                                                      |
| `index`                         | Stores the topic number (0, 1, 2, ...).                                                                              |
| `topic`                         | Contains the weight (importance) of every word for the current topic.                                                |
| `enumerate()`                   | Returns both the topic number and topic data.                                                                        |
| `nmf_model.components_`         | A matrix where each row represents a discovered topic and each column represents the weight of a word in that topic. |
| `results =`                     | Creates a list to store the top words of the current topic.                                                          |
| `tfidf.get_feature_names_out()` | Retrieves all words (features) learned by the TF-IDF vectorizer.                                                     |
| `topic.argsort()`               | Sorts the word weights from lowest to highest.                                                                       |
| `[-10:]`                        | Selects the indices of the 10 words with the highest weights.                                                        |
| `[i]`                           | Retrieves the corresponding word using its index.                                                                    |
| `print(results)`                | Prints the top 10 important words for the current topic.                                                             |


This line calculates how strongly each chat message belongs to every topic.

```python
topic_results = nmf_model.transform(dtm)
```

| Code            | Description                                                                                       |
| --------------- | ------------------------------------------------------------------------------------------------- |
| `topic_results` | Variable that stores the topic probability/weight for every document (chat message).              |
| `=`             | Assigns the output of the function to the variable.                                               |
| `nmf_model`     | The trained NMF Topic Modeling model.                                                             |
| `.transform()`  | Applies the trained NMF model to the TF-IDF matrix and calculates topic weights for each message. |
| `dtm`           | Document-Term Matrix (TF-IDF representation of chat messages).                                    |


Assigns one topic number (the most dominant topic) to each chat message.

```python
df["Topic"] = topic_results.argmax(axis=1)
```

| Code            | Description                                            |
| --------------- | ------------------------------------------------------ |
| `df["Topic"]`   | Creates a new column named **Topic** in the DataFrame. |
| `=`             | Assigns the predicted topic to each message.           |
| `topic_results` | Matrix containing topic weights for every message.     |
| `.argmax()`     | Finds the index of the highest topic weight.           |
| `axis=1`        | Performs the search row-wise (for each message).       |
